# Activation function investigation


````python
import numpy as np
import math

from numpy import load
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import load_model
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras. layers import Conv2D, MaxPooling2D
from keras import backend as K
from keras import regularizers

import tensorflow  as tf
from keras.layers import *
from keras.callbacks import ReduceLROnPlateau

import matplotlib.pyplot as plt

from timeit import default_timer as timer
````

### Introduction
In this file I will run neural networks for the 15 class plant disease classification problem on the downsampled subset of the plant village data. The activation function will be varied, and I will compare the performance based on the accuracy and loss curves on the validation set. I will also time the training of each neural network to gain insight into the scalablity, as complexity is a particularly important consideration in situations with large data.

### Activation functions

The four activation functions I will compare are Sigmoid, Tanh, ReLu and Swish. These are defined and plotted in the code chunk below. 

````python
# Sigmoid Function
def sigmoid(x):
    a = []
    for i in x:
        a.append(1/(1+math.exp(-i)))
    return a


# Hyperbolic Tanjent Function
def tanh(x, derivative=False):
    if (derivative == True):
        return (1 - (x ** 2))
    return np.tanh(x)


# ReLU Function
def re(x):
    b = []
    for i in x:
        if i<0:
            b.append(0)
        else:
            b.append(i)
    return b


# Intervals for graph
x = np.arange(-3., 3., 0.1)
sig = sigmoid(x)
tanh = tanh(x)
relu = re(x)
swish = sig*x


# Displaying the functions
line_1, = plt.plot(x,sig, label='Sigmoid')
line_2, = plt.plot(x,tanh, label='Tanh')
line_3, = plt.plot(x,relu, label='ReLU')
line_4, = plt.plot(x,swish, label='Swish')
plt.legend(handles=[line_1, line_2, line_3, line_4])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
````

### Loading the data

````python
# load numpy array from npy file
x_train = load('./x_train_sample.npy')
# check import
print(x_train[0])
````

````python
x_test = load('./x_test_sample.npy')
y_train = load('./y_train_sample.npy')
y_test = load('./y_test_sample.npy')
````

### Fitting models

The architecture for the neural network will be the same for each neural network, and only the activation function will be varied. The model will contain two convolutional layers, and we will use the softmax activation function for the output layer, as it returns the probability for a datapoint belonging to each individual class and so is well suited for multiclass classification problems. A diagram of the architecture is plotted below the model training.


````python
batch_size = 128 # Batch-Size
num_classes = 15 # Number of classes
epochs = 20 # Number of epochs
w_l2 = 1e-5 
````

````python
# This reduces the learning rate when the model is no longer improving in terms of the validation loss metric
# Prevents overfitting.
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-5)
````

````python
input_shape = (32, 32, 3) # shape of input 
````

### Sigmoid

````python
#Define model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),  kernel_regularizer=regularizers.l2(w_l2),
                 input_shape=input_shape))
model.add(BatchNormalization()) # batch normalisation - helps to reduce the impact of vanishing gradients and exploding gradients
model.add(Activation('sigmoid'))
model.add(Conv2D(64, (3, 3),  kernel_regularizer=regularizers.l2(w_l2)))
model.add(BatchNormalization())
model.add(Activation('sigmoid')) 
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, kernel_regularizer=regularizers.l2(w_l2)))
model.add(BatchNormalization())
model.add(Activation('sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
model.summary()
````

````python
start = timer()
hist_sigmoid=model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test), callbacks=[reduce_lr])
end = timer()

score = model.evaluate(x_test, y_test, verbose=0)


print('Test loss:', score[0])
print('Test accuracy:', score[1])

time_sigmoid = end - start 
print('Time to run', time_sigmoid, ' seconds')
````

### Hyperbolic tangent

````python
#Defining the model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),  kernel_regularizer=regularizers.l2(w_l2),
                 input_shape=input_shape))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(Conv2D(64, (3, 3),  kernel_regularizer=regularizers.l2(w_l2)))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, kernel_regularizer=regularizers.l2(w_l2)))
model.add(BatchNormalization())
model.add(Activation('tanh'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
model.summary()
````

````python
start = timer()

hist_tanh=model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test), callbacks=[reduce_lr])

end = timer()

score = model.evaluate(x_test, y_test, verbose=0)

print('Test loss:', score[0])
print('Test accuracy:', score[1])

time_tanh = end - start 
print('Time to run', time_tanh, ' seconds')
````

### ReLu

````python
# Defining the model 
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),  kernel_regularizer=regularizers.l2(w_l2),
                 input_shape=input_shape))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3),  kernel_regularizer=regularizers.l2(w_l2)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, kernel_regularizer=regularizers.l2(w_l2)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
model.summary()
````

````python
start = timer()

hist_relu=model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test), callbacks=[reduce_lr])
end = timer()

score = model.evaluate(x_test, y_test, verbose=0)

print('Test loss:', score[0])
print('Test accuracy:', score[1])

time_relu = end - start 
print('Time to run', time_relu, ' seconds')
````

````python

def swish(x):
    return x * K.sigmoid(x)
    # return x * (1/(1 + K.exp(-x)))
````

````python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), kernel_regularizer=regularizers.l2(w_l2),
                 input_shape=input_shape, kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Activation(swish))
model.add(Conv2D(64, (3, 3), kernel_regularizer=regularizers.l2(w_l2), kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Activation(swish))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, kernel_regularizer=regularizers.l2(w_l2), kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Activation(swish))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])

model.summary()
````

````python
start = timer()

hist_swish=model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test), callbacks=[reduce_lr])

end = timer()

score = model.evaluate(x_test, y_test, verbose=0)

print('Test loss:', score[0])
print('Test accuracy:', score[1])

time_swish = end - start 
print('Time to run', time_swish, ' seconds')
````

### Results

Now that the models are trained, I move on to plotting the results. First, we define a plotting function for the accuracy and loss metrics, which will exhibit the performance of the models over time.

````python
def plot_history(hists, attribute='val_loss', axis=(-1,21,0.85,0.94), loc='lower right'):
    title = {'val_loss': 'valid. loss', 'loss': 'trn. loss', 'val_accuracy': 'valid. accuracy', 'accuracy': 'trn. accuracy'}
    num_hists = len(hists)
    
    plt.figure(figsize=(12, 8))  
    plt.axis(axis)
    for i in range(num_hists):
        plt.plot(hists[i].history[attribute])
    plt.title(title[attribute])  
    plt.ylabel(title[attribute])  
    plt.xlabel('epoch')  
    plt.legend(['Sigmoid', 'tanh', 'ReLU', 'Swish'], loc=loc)  

    plt.show()
````

````python
hists = [hist_sigmoid, hist_tanh, hist_relu, hist_swish] # list of histories from the models
````

````python
history_dict = hist_sigmoid.history #check the keys 
print(history_dict.keys())
````

### Accuracy

````python
# Validation accuracy
plot_history(hists, attribute='val_accuracy',axis=(-1,21,0.0,1.2), loc='lower right')
````

````python
# train accuracy
plot_history(hists, attribute='accuracy', axis=(-1,21,0.0,1.2), loc='lower right')
````

The training accuracy curve shows how the models using tanh, ReLu and Swish were all able to reach almost 100%, however the model using sigmoid had a significantly lower learning rate, and attained a far lower training accuracy.

In the validation accuracy curve, again ReLu, Swish and tanh produce similar results, with ReLu exhibiting a slightly higher performance towards the end. The model using sigmoid has a very low accuracy until epoch 10, when the accuracy begins to climb, but does not quite reach the accuracy of the other three models. The loss curves mirror these results.

### Efficiency

We now investigate the time taken for the training of each model.

````python
times = [time_sigmoid, time_tanh, time_relu, time_swish]
````

````python
final_val_accuracy = []

for i in range(4):
    final_val_accuracy.append(hists[i].history['val_accuracy'][19]) #retrieve accuracy 
    
print(final_val_accuracy)
````

````python
n = ['Sigmoid', 'Tanh', 'Relu', 'Swish'] 

fig, ax = plt.subplots()
ax.scatter(times, final_val_accuracy)

for i, txt in enumerate(n):
    ax.annotate(txt, (times[i], final_val_accuracy[i]))
    
plt.title("Final Validation accuracy and runtime")
plt.xlabel("Time (seconds)")
plt.ylabel("Validation accuracy")
````

The above plot indicates that whilst the Swish model attains a similar accuracy to tanh and ReLu, it is over 10% slower. In this study we only used a very small sample dataset, and so in the case of much larger data, this difference would likely be magnified. Hence we can conclude that in this case, ReLu or Tanh may be better choices. The sigmoid model yields a lower accuracy and is slower than Tanh and Relu and so it doesn't seem to be the correct choice for this problem. This aligns with the suggestion in ```[2]``` that the Sigmoid activation function performs best in binary classification problems, whilst Tanh is better for multi class problems. ```[2]``` also suggests that ReLu is the best general choice for activation function, and this seems to be a good suggestion in line with the results of this experiment, as ReLu achieved the highest performance in the least amount of time.
